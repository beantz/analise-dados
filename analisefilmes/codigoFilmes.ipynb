```python
import pandas as pd
df = pd.read_csv("archive/IMDbMovies.csv", index_col = 0)
#corrigir as linhas que possuem valores vazios
tabela = df.dropna()
tabela
```

```python
#FILMES QUE MAIS VENDERAM BEM
#Selecionando apenas filmes do seculo 21
filmesSeculo21 = tabela[tabela['Release Year'] > 2001]
#Filmes que mais venderam bem no seculo 21
venderamBem = filmesSeculo21.sort_values(by=['Gross worldwide'], ascending=False)
venderamBem['Gross worldwide'] = venderamBem['Gross worldwide'].replace(r'[\$,]', '', regex=True).astype(float)
# FunÃ§Ã£o para formatar dinheiro
def format_money(money):
    return '${:,.2f}'.format(money)
# Aplicar formataÃ§Ã£o Ã  coluna 'Gross worldwide'
venderamBem['Gross worldwide'] = venderamBem['Gross worldwide'].apply(format_money)
res = venderamBem['Gross worldwide'].head(20)
res
```

```python
#OS 20 FILMES QUE MENOS VENDERAM
venderamMenos = filmesSeculo21.sort_values(['Gross worldwide'], ascending=True)
def format_money(money):
    return '${:,.2f}'.format(money)

venderamMenos['Gross worldwide'] = venderamMenos['Gross worldwide'].apply(format_money)
res = venderamMenos['Gross worldwide'].head(20)
res
```

```python
#FILMES COM MAIS LONGA DURAÃ‡ÃƒO
# Criando o DataFrame
df = pd.DataFrame(filmesSeculo21)

# FunÃ§Ã£o para converter 'Runtime' para minutos
def convert_to_minutes(runtime):
    parts = runtime.split()
    hours = int(parts[0][:-1]) * 60  # Remove o 'h' e converte para minutos
    minutes = int(parts[1][:-1]) if len(parts) > 1 else 0  # Remove o 'm' e converte para minutos
    return hours + minutes

# Aplicando a funÃ§Ã£o de conversÃ£o
df['Runtime (minutes)'] = df['Runtime'].apply(convert_to_minutes)

# Ordenando os filmes pelo tempo de execuÃ§Ã£o
df_sorted = df.sort_values(by='Runtime (minutes)', ascending=False)

# Selecionando os tÃ­tulos dos filmes mais longos junto com seus tempos de execuÃ§Ã£o
longest_movies = df_sorted[['Runtime']]

# Exibindo os resultados
print(longest_movies)
```

```python
#FILMES COM AS MELHORES AVALIAÃ‡Ã•ES DENTRE OS MAIS CAROS(valor acima de 1.5 bilhÃµes)
# Garantir que a coluna 'Rating' estÃ¡ no formato string
filmesSeculo21['Rating'] = filmesSeculo21['Rating'].astype(str)

# Extrair a parte numÃ©rica das avaliaÃ§Ãµes e converter para float
filmesSeculo21['Rating'] = pd.to_numeric(filmesSeculo21['Rating'].str.extract(r'(\d+\.\d+)')[0], errors='coerce')

# Remover o sÃ­mbolo de dÃ³lar e vÃ­rgulas e converter 'Gross worldwide' para float
filmesSeculo21['Gross worldwide'] = filmesSeculo21['Gross worldwide'].replace('[\$,]', '', regex=True).astype(float)

# Definir o critÃ©rio para os filmes mais caros (exemplo: filmes com gross worldwide > 1.5 bilhÃµes)
criterio = 1_000_000_000.00
filmes_mais_caros = filmesSeculo21[filmesSeculo21['Gross worldwide'] > criterio]
# Selecionar apenas as colunas 'Gross worldwide' e 'Rating' dos filmes mais caros
colunas_selecionadas = filmes_mais_caros[['Rating']]

# Exibir as colunas selecionadas
print(colunas_selecionadas)
```

```python
#DENTRE OS QUE MAIS VENDERAM BEM, PEGANDO APENAS OS COM AVALLIAÃ‡ÃƒO MAIS BAIXA
colunas_selecionadas = filmes_mais_caros[['Rating']]

colunas_selecionadas = colunas_selecionadas.sort_values(by='Rating').head(23)

print(colunas_selecionadas)
```

```python
#FILMES MAIS CAROS NO CANADA E US
# Definir o critÃ©rio para os filmes mais caros (exemplo: filmes com gross worldwide > 1 bilhÃ£o)
criterio = 1_000_000_000.00
filmes_mais_caros = filmesSeculo21[filmesSeculo21['Gross worldwide'] > criterio]

# Selecionar apenas a coluna 'Gross in US & Canada' dos filmes mais caros
gross_us_canada = filmes_mais_caros['Gross in US & Canada'].head(15)
print(gross_us_canada)
```

```python
# Contar a frequÃªncia dos gÃªneros
genre_counts = filmesSeculo21['Main Genres'].str.split(',', expand=True).stack().value_counts().head(5)

# Exibir os gÃªneros mais frequentes
print(genre_counts)
```

```python
#TOP 15 DIRETORES QUE MAIS LANÃ‡ARAM FILMES NO SECULO 21
# Contar a frequÃªncia dos diretores
director_counts = filmesSeculo21['Director'].value_counts().head(15)

# Exibir os diretores mais frequentes
print(director_counts)
```

```python
 
# Remover caracteres nÃ£o numÃ©ricos do orÃ§amento
filmesSeculo21['Budget'] = filmesSeculo21['Budget'].replace('[\$,]', '', regex=True)

# Confirmar que a coluna Budget Ã© tratada como string
filmesSeculo21['Budget'] = filmesSeculo21['Budget'].astype(str)

# Extrair apenas os nÃºmeros (incluindo decimais) da string
filmesSeculo21['Budget'] = filmesSeculo21['Budget'].str.extract(r'(\d+\.?\d*)').astype(float)

# Ordenar os filmes pelo orÃ§amento de forma decrescente e exibir os top 15
mais_caros_produzir = filmesSeculo21.sort_values(by='Budget', ascending=False).head(15)

# Exibir os filmes que mais tiveram um orÃ§amento alto
print(mais_caros_produzir[['Budget']])
```
